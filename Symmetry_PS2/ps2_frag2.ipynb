```python
import csv
import matplotlib.pyplot as plt

# Specify the path to your CSV file
csv_file_path = '../problems/frag2_sol.csv'

# Initialize lists to store coordinates
x1 = []
y1 = []
x2 = []
y2 = []

# Read the CSV file
with open(csv_file_path, mode='r', newline='') as file:
    csv_reader = csv.reader(file)
    
    # Iterate through the rows and extract coordinates
    for row in csv_reader:
        x1.append(float(row[0]))  # First column as x1
        y1.append(float(row[1]))  # Second column as y1
        x2.append(float(row[2]))  # Third column as x2
        y2.append(float(row[3]))  # Fourth column as y2

# Plot the coordinates
plt.figure(figsize=(10, 6))
plt.scatter(x1, y1, color='blue', label='Coordinates (x1, y1)', alpha=0.5)
plt.scatter(x2, y2, color='red', label='Coordinates (x2, y2)', alpha=0.9)

# Optionally, connect the points with lines
plt.plot(x1, y1, color='blue', alpha=0.3)
plt.plot(x2, y2, color='red', alpha=0.8)

# Add labels and title
plt.title('Scatter Plot of Coordinates from CSV')
plt.xlabel('X Coordinates')
plt.ylabel('Y Coordinates')
plt.legend()
plt.grid()

# Show the plot
plt.axis('equal')
plt.show()
```

```python
# input
shape = list(zip(x1, y1, x2, y2)) 
```

```python
# Convert to set for easier lookup
original_points = set((x, y) for x, y, _, _ in shape)  # Only x, y pairs

# Calculate the midpoint for vertical symmetry
x_coords = [x for x, y, _, _ in shape]
midpoint = sum(x_coords) / len(x_coords)
```

```python
# Check for vertical symmetry
is_symmetric = True
for x, y, _, _ in shape:
    # Calculate the reflected point across the vertical line (midpoint)
    reflected_point = (2 * midpoint - x, y)
    if reflected_point not in original_points:
        is_symmetric = False
        break

# Output the result
if is_symmetric:
    print("The shape has vertical symmetry.")
else:
    print("The shape does not have vertical symmetry.")
```

```python
# Check for horizontal symmetry
is_symmetric = True
for x, y, _, _ in shape:
    # Calculate the reflected point across the horizontal line (midpoint)
    reflected_point = (x, 2 * midpoint - y)
    if reflected_point not in original_points:
        is_symmetric = False
        break

# Output the result
if is_symmetric:
    print("The shape has horizontal symmetry.")
else:
    print("The shape does not have horizontal symmetry.")
```

```python
# Check for diagonal symmetry with respect to the line y = x
is_symmetric = True
for x, y, _, _ in shape:
    # Calculate the reflected point across the line y = x
    reflected_point = (y, x)  # Swap x and y for reflection across y = x
    if reflected_point not in original_points:
        is_symmetric = False
        break

# Output the result
if is_symmetric:
    print("The shape has diagonal symmetry with respect to the line y = -x.")
else:
    print("The shape does not have diagonal symmetry with respect to the line y = -x.")
```

```python
# Check for diagonal symmetry with respect to the line y = x
is_symmetric = True
for x, y, _, _ in shape:
    # Calculate the reflected point across the line y = x
    reflected_point = (x, y)  # Swap x and y for reflection across y = x
    if reflected_point not in original_points:
        is_symmetric = False
        break

# Output the result
if is_symmetric:
    print("The shape has diagonal symmetry with respect to the line x = y.")
else:
    print("The shape does not have diagonal symmetry with respect to the line x = y.")
```

```python
control_points_array = shape

# Convert to set for easier lookup
original_points = set((x, y) for x, y, _, _ in control_points_array)

# Check for reflectional symmetry about the y-axis
reflection_symmetric = True
for x, y in original_points:
    if (-x, y) not in original_points:
        reflection_symmetric = False
        break

# Check for rotational symmetry (example: 180 degrees around the origin)
rotational_symmetric = True
for x, y in original_points:
    if (-x, -y) not in original_points:
        rotational_symmetric = False
        break

# Output the results
if reflection_symmetric:
    print("The shape has reflectional symmetry about the y-axis.")
else:
    print("The shape does not have reflectional symmetry about the y-axis.")

if rotational_symmetric:
    print("The shape has rotational symmetry about the origin.")
else:
    print("The shape does not have rotational symmetry about the origin.")
```
